# Read Resume PDF
Load the checked-in `Sumeet Resume.pdf` file and extract its text using PyPDF2.


```python
from pathlib import Path
import PyPDF2
pdf_path = Path("Sumeet Resume.pdf")
assert pdf_path.exists(), f"Resume PDF not found at {pdf_path}"
reader = PyPDF2.PdfReader(str(pdf_path))
resume_text = "
".join(page.extract_text() or "" for page in reader.pages)
print(resume_text[:1200])
```

# Extract Resume Details
Parse the resume text for the title, summary, skills, experience, and education.


```python
lines = [line.strip() for line in resume_text.splitlines() if line.strip()]
for i, line in enumerate(lines[:80], start=1):
    print(f"{i:02}: {line}")
```
